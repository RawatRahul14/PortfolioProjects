```python
# import all the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import figure
import warnings

warnings.simplefilter("ignore")
```

```python
# import the required csv file

df = pd.read_csv("hr.csv")
df.head()
```

```python
# to check the data types and the counts of all the columns

df.info()
```

```python
# making a correlation table between all the numerical values

corr = df.corr()
```

```python
plt.figure(figsize= (25, 25))
sns.heatmap(corr, annot = True);
```

```python
# finding out the percentage of people doing overtime work

print(df["OverTime"].value_counts())

plt.figure(figsize= (6, 6))
df["OverTime"].value_counts().plot(kind= "pie")
plt.title("Percentage of people doing OverTime");
```

```python
# all the different types of the marital status in the dataset

df["MaritalStatus"].unique()
```

```python
# plotting a pie chart of the different types of the marital status

df["MaritalStatus"].value_counts().plot(kind= "pie", figsize= (6, 6))
plt.title("Marital Status")
```

```python
# counting the values of different job titles

job = df["JobRole"].value_counts()
```

```python
# plotting a bar graph of the number of people doing the different types of the jobs

plt.figure(figsize= (14, 6))
sns.barplot(job, job.index)
plt.xlabel("Number of people")
plt.ylabel("Different Job Roles");
```

```python
# making a pie chart on the basis of the gender

df["Gender"].value_counts().plot(kind= "pie", figsize= (6, 6))
plt.title("Percentage of different Genders");
```

```python
# number of people with the field education

field= df["EducationField"].value_counts()
```

```python
plt.figure(figsize= (14, 6))
sns.barplot(field, field.index)
plt.xlabel("number of people")
plt.ylabel("Education field");
```

```python
# ratio of the people doing jobs in different departments

df["Department"].value_counts().plot(kind= "pie", figsize= (6, 6))
plt.title("Ratio of people working in Different departments");
```

```python
# number of people doing the business travels

df["BusinessTravel"].value_counts().plot(kind= "pie", figsize= (6, 6))
plt.title("Ratio of people who goes on Business trips");
```

```python
# plotting a bar chart based on the number of people with the total working hours they been doing

cond = df["TotalWorkingYears"].value_counts().sort_values(ascending= False)
plt.figure(figsize= (14, 6))
cond.plot(kind= "bar")
plt.xlabel("Working hours")
plt.ylabel("Number of people");
```

```python
plt.figure(figsize= (6, 6))
df["Education"].value_counts().plot(kind= "pie");
plt.title("Education")
```

```python
# number of companies people had previously worked in 

df["NumCompaniesWorked"].value_counts().plot(kind= "bar", figsize= (12, 5))
plt.xlabel("Total number of companies people had worked")
plt.ylabel("Number of people");
```

```python
# top 8 

df["DistanceFromHome"].value_counts().nlargest(8)

# average distance from home to work place

print("Average distance of people to their work place is {0}".format(df["DistanceFromHome"].mean()))
```

```python
df["DistanceFromHome"].value_counts().plot(kind= "bar", figsize= (14, 6));
plt.xlabel("Distance in Kms")
plt.ylabel("Number of people")
```
